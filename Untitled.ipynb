```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
image = cv2.imread('lines.png', cv2.IMREAD_GRAYSCALE)

horizontal_mask = np.array([[-1, -1, -1],
                             [ 2,  2,  2],
                             [-1, -1, -1]])
vertical_mask = np.array([[-1,  2, -1],
                          [-1,  2, -1],
                          [-1,  2, -1]])
diagonal_mask = np.array([[ 2, -1, -1],
                          [-1,  2, -1],
                          [-1, -1,  2]])

horizontal_edges = cv2.filter2D(image, -1, horizontal_mask)
vertical_edges = cv2.filter2D(image, -1, vertical_mask)
diagonal_edges = cv2.filter2D(image, -1, diagonal_mask)

plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1), plt.imshow(horizontal_edges, cmap='gray'), plt.title('Horizontal')
plt.subplot(1, 3, 2), plt.imshow(vertical_edges, cmap='gray'), plt.title('Vertical')
plt.subplot(1, 3, 3), plt.imshow(diagonal_edges, cmap='gray'), plt.title('Diagonal')
plt.show()
```

```python
image2 = cv2.imread('lion.png', cv2.IMREAD_GRAYSCALE)

sobel_x = cv2.Sobel(image2, cv2.CV_64F, 1, 0, ksize=3) 
sobel_y = cv2.Sobel(image2, cv2.CV_64F, 0, 1, ksize=3) 

combined_edges = cv2.magnitude(sobel_x, sobel_y)

plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1), plt.imshow(np.abs(sobel_x), cmap='gray'), plt.title('Horizontal Edges')
plt.subplot(1, 3, 2), plt.imshow(np.abs(sobel_y), cmap='gray'), plt.title('Vertical Edges')
plt.subplot(1, 3, 3), plt.imshow(np.abs(combined_edges), cmap='gray'), plt.title('Combined Edges')
plt.show()
```

```python
coin_image = cv2.imread('coin.png', cv2.IMREAD_GRAYSCALE)

blurred = cv2.GaussianBlur(coin_image, (5, 5), 0)

edges = cv2.Canny(blurred, 50, 150)

plt.figure(figsize=(8, 6))
plt.subplot(1, 2, 1), plt.imshow(coin_image, cmap='gray'), plt.title('Original Coin Image')
plt.subplot(1, 2, 2), plt.imshow(edges, cmap='gray'), plt.title('Edge Detection Output')
plt.show()
```
